# Utility functions

```python
def digit_to_int(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [ digit_to_int(c) for c in re.split(r'(\d+)', text) ]
```

# Get All JSON Files

Sorting the json files is important because the data is supposed to be organised in chronological order. However, sorting without keys doesn't yield proper results. The sorted json files are in this order:

['wz-stats-1.json',
'wz-stats-10.json',
'wz-stats-11.json',
'wz-stats-12.json',
...,
'wz-stats-2.json',
'wz-stats-20.json',
... ]

For this reason we use the Utility function **natural_keys()**

```python
json_files = [f for f in listdir(api_files_base_path) if f.endswith(".json")]
json_files.sort(key=natural_keys)
```

# Normalize and Append all JSON Files in a List

The json data has a hierarchical and we need to flatten this out. For this purpose we will the **json_normalize()** function in the Pandas library. We will be process each file individually and then append the resulting dataframe into a list.

```python
for file in json_files:
    with open(api_files_base_path+file, 'r') as jf:
        json_data = json.load(jf)
    
    match_data = json_data['data']['matches']
    df = pd.json_normalize(match_data, record_path=['segments'], meta=meta_columns, errors='ignore')
    df_list.append(df)
```

# Flatten The Teammates Column

The "metadata.teammates" columns contains a list of teammates, we need to flatten this out. We will create a new column for each teammate.

```python
for row_index, row in df_copy.iterrows():
    if row['metadata.teammates'] != None and len(row['metadata.teammates']) > 0:
        teammates = row['metadata.teammates']
        for idx, mate in enumerate(teammates):
            teammate_nmr = idx + 1
            mate_name = 'teammate_'+ str(teammate_nmr) +'.name'
            mate_kills = 'teammate_'+ str(teammate_nmr) +'.kills'
            mate_deaths = 'teammate_'+ str(teammate_nmr) +'.deaths'
            mate_placement = 'teammate_'+ str(teammate_nmr) +'.placement'
            mate_time_played = 'teammate_'+ str(teammate_nmr) +'.time_played'

            df_copy.at[row_index, mate_name] = mate['platformUserHandle']
            df_copy.at[row_index, mate_kills] = mate['stats']['kills'] 
            df_copy.at[row_index, mate_deaths] = mate['stats']['deaths']
            df_copy.at[row_index, mate_placement] = mate['stats']['placement']
            df_copy.at[row_index, mate_time_played] = mate['stats']['timePlayed']
```

# Utility functions

```python
def digit_to_int(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [ digit_to_int(c) for c in re.split(r'(\d+)', text) ]
```

# Format & Extract Data

```python
def get_my_stats(stats, my_name):
    my_stats = {}

    interesting_stats = ['kills', 'deaths', 'assists', 'kdRatio', 'placement', 'damageDone', 'damageTaken', 'percentTimeMoving',
                         'objectiveBrDownEnemyCircle1', 'objectiveBrDownEnemyCircle2', 'objectiveBrDownEnemyCircle3',
                        'objectiveBrDownEnemyCircle4', 'objectiveBrDownEnemyCircle5', 'distanceTraveled']

    for stat in interesting_stats:
        my_stats[my_name + '_' + stat] = stats.get(stat, {}).get('value')

    return my_stats
```

```python
def get_teammates_stats(team_mates):
    team_mates_stats = {}

    interesting_stats = ['kills', 'deaths']
    
    for team_mate in (team_mates or []):
        team_mate_name = team_mate.get('platformUserHandle', '').split("#")[0]
        team_mate_stats = {f'{team_mate_name}_{stat}': team_mate.get('stats', {}).get(stat) for stat in interesting_stats}
        team_mates_stats.update(team_mate_stats)

    return team_mates_stats
```

```python
def extract_data(match):
    match_data = {}

    match_data['match_id'] = match.get('attributes', {}).get('id')
    
    metadata = match.get('metadata', {})
    match_data['player_count'] = metadata.get('playerCount')
    match_data['team_count'] = metadata.get('teamCount')
    match_data['map_name'] = metadata.get('mapName')
    match_data['mode_name'] = metadata.get('modeName')
    match_data['timestamp'] = metadata.get('timestamp')
    match_data['duration_value'] = metadata.get('duration', {}).get('value')
    match_data['duration_display_value'] = metadata.get('duration', {}).get('displayValue')
    
    segment = match['segments'][0]
    # Get my stats
    my_name = 'rfy'
    my_stats = get_my_stats(segment.get('stats', {}), my_name)
    match_data.update(my_stats)
    
    team_mates_stats = get_teammates_stats(segment.get('metadata', {}).get('teammates'))
    match_data.update(team_mates_stats)
    
    return match_data
```

```python
def format_match_data(match_data):
    match_data = [extract_data(match) for match in match_data]
    return match_data
```

# Normalize and Append all JSON Files in a List

The json data has a hierarchical and we need to flatten this out. For this purpose we will the **json_normalize()** function in the Pandas library. We will be process each file individually and then append the resulting dataframe into a list.

```python
for file in json_files:
    json_data = None
    with open(api_files_base_path+file, 'r') as jf:
        json_data = json.load(jf)
    
    match_data = json_data['data']['matches']
    data = format_match_data(match_data)
    
    df = pd.DataFrame(data)
    # df = pd.json_normalize(match_data, record_path=['segments'], meta=meta_columns, errors='ignore')
    DF_LIST.append(df)
```
